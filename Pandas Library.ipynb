```python
#creating new data and apply some properties
data1 = pd.DataFrame({"Key1" : [1, 2, 4, 5, 6],
                      "Key2" : [4, 5, 6, 7, 8],
                      "Key3" : [3, 4, 5, 6, 6]})
data1
```

```python
data2 = pd.DataFrame({"Key1" : [1, 2, 45, 6, 67],
                      "Key4" : [56, 5, 6, 7, 8],
                      "Key5" : [3, 56, 5, 6, 6]})
data2
```

```python
#merging two tables withs same order
pd.merge(data1, data2) #here, tables are merged on basis of 'inner' join(by default)
                        #where, key1 is common on both tables so it create the common column table
```

```python
pd.merge(data1, data2, how = "left") #here, tables are merged on basis of 'left' join
                                    #where, table focused on all data of left table ie. data1
```

```python
pd.merge(data1, data2, how = "right") #here, tables are merged on basis of 'right' join
                                    #where, table focused on all data of right table ie. data2
```

```python
pd.merge(data1, data2, how = "outer") #here, table is union of (left join + right join)
```

```python
pd.merge(data1, data2, how = "cross")  #here, table is cross product of all the rows of table with each row of anther column
```

```python
#creating the column which categories the 'fare' on basis of 'cheap', 'mid' and 'high' value of standards
def categ_fare(x):
    if x < 10 :
        return "Cheap"
    elif x >= 10 and x < 20:
        return "Mid"
    else:
        return "High"
```

```python
df["categ_fare"] = df["Fare"].apply(categ_fare)
```
