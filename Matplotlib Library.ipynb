```python
#importing the libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.title("This is sinusoidal graph")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
colours = np.random.rand(50)
plt.scatter(x, y, c = colours)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
sizes = 500 * np.random.rand(50)
colours = np.random.rand(50)
plt.scatter(x, y, c = colours, sizes = sizes)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
sizes = 500 * np.random.rand(50)
colours = np.random.rand(50)
plt.scatter(x, y, c = colours, sizes = sizes, alpha = .5)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot")
plt.show()
```
