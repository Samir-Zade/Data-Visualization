```python
#loading tips dataset by seaborn
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
#try to visualize who pay more tips compared to bill paid by them
sns.scatterplot(x = tips.total_bill, y = tips.tip)
plt.show()
```

```python
tips.smoker.value_counts()
```

```python
#relplot = relational plot between x and y
sns.relplot(x = tips.total_bill, y = tips.tip, data = tips, hue = "smoker") 
plt.show()
```

```python
#colour on basis of size, but it's not visualize well so we wil used next feature called 'style'.
sns.relplot(x = tips.total_bill, y = tips.tip, data = tips, hue = "size") 
plt.show()
```

```python
tips["size"].value_counts()
```

```python
#using style instead of hue for more distinct way of referencing the types of data
sns.relplot(x = tips.total_bill, y = tips.tip, data = tips, style = "size") 
plt.show()
```

```python
#using 'hue' with 'style' for better visualization
sns.relplot(x = tips.total_bill, y = tips.tip, data = tips, style = "size", hue = "time") 
plt.show()
```

```python
tips.head()
```

```python
#try to visualize the respective days when the no of customer increases with the time of taking food ie. dinner or lunch
sns.catplot(x = "day", y = "total_bill", data = tips, hue = "time")
plt.show()
```

```python
#visulizing does smoker pay more tip or non smoker pay's more with their gender
sns.catplot(x = "smoker", y = "tip", data = tips, hue = "sex")
plt.show()
```

```python
#"jointplot()" joints two diffrent axis and shows it's scatter plot and histogram at same time.
sns.jointplot(x = "total_bill", y = "tip", data = tips)
plt.show()
```

```python
#pairplot shows the relation between all the column's data present in dataset
sns.pairplot(tips) 
plt.show()
```
